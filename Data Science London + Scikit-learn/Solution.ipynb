```python
import pandas as pd
import numpy as np
import warnings
from matplotlib import style, pyplot as plt
style.use('ggplot')
#warnings.filterwarnings('ignore')
```

```python
df_train = pd.read_csv('train.csv',header=None)
df_label = pd.read_csv('trainLabels.csv',header=None)
df_test = pd.read_csv('test.csv',header=None)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_validation,Y_train,Y_validation = train_test_split(df_train,df_label,test_size=0.2)
print(X_train.shape,X_validation.shape,Y_train.shape,Y_validation.shape)
```

```python
from xgboost import XGBClassifier
clf = XGBClassifier(max_depth = 11,learning_rate=0.05,n_estimators=100,silent=False)
clf.fit(X_train,Y_train)
clf.score(X_validation,Y_validation)
```

```python
from lightgbm import LGBMClassifier
clf = LGBMClassifier(num_leaves=20,
                     max_depth=-1,
                     learning_rate=0.4,
                     n_estimators=70,
                     silent=False,
                     objective="binary")
clf.fit(X_train,Y_train,eval_set=(X_validation,Y_validation))
clf.score(X_validation,Y_validation)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
param_grid= {
    'leaf_size':[5,10,30,50],
}
gridsearch_knn = GridSearchCV(knn,param_grid=param_grid,scoring='accuracy',verbose=5,n_jobs=-1,cv=5)
gridsearch_knn.fit(df_train,df_label)
print(gridsearch_knn.best_estimator_)
best_knn = gridsearch_knn.best_estimator_
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
param_grid = {
    'n_estimators':[5,10,50],
    'max_depth':[3,5,10,None]
}
gridsearch_rf = GridSearchCV(rf,param_grid=param_grid,scoring='accuracy',cv=5,verbose=3)
gridsearch_rf.fit(df_train,df_label)
print(gridsearch_rf.best_estimator_)
best_rf = gridsearch_rf.best_estimator_
```

```python
from sklearn.ensemble import VotingClassifier
clf = VotingClassifier([('knn',best_knn),('rf',best_rf)])
clf.fit(df_train,df_label)
prediction = clf.predict(df_test)
print(clf.score(df_train,df_label))
len(prediction)
```

```python
from sklearn.svm import SVC
clf = SVC()
clf.fit(df_train,df_label)
prediction = clf.predict(df_test)
print(clf.score(df_train,df_label))
len(prediction)
```

```python
res =  np.column_stack((range(1,9001),prediction))
ans =pd.DataFrame(data=res,columns=['Id','Solution'])
ans.to_csv('sol.csv',index=False)
ans
```
