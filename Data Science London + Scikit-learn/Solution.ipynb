```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df_train = pd.read_csv('train.csv',header=None)
df_label = pd.read_csv('trainLabels.csv',header=None)
df_test = pd.read_csv('test.csv',header=None)
```

```python
from sklearn.svm import SVC
clf = SVC()
clf.fit(df_train, df_label)
clf.score(df_train, df_label)
```

```python
prediction = clf.predict(df_test)
len(prediction)
```

```python
res =  np.column_stack((range(1,9001),prediction))
ans =pd.DataFrame(data=res,columns=['Id','Solution'])
ans.to_csv('sol.csv',index=False)
ans
```
